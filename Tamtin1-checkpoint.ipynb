# تمرین 3


```python
my_list = [1, 2, 3, 4, 2, 3, 5, 6, 1, 7]

my_set = set(my_list)

occurrences = {}
for item in my_list:
    if item in occurrences:
        occurrences[item] += 1
    else: occurrences[item] = 1

print("Original List:", my_list)
print("Converted Set:", my_set)
print("Occurrences of elements in the list:")
for key, value in occurrences.items():
    print(f"{key}: {value} times")
```

# تمرین 8

```python
def unique_characters(file_name):
    try:
        with open(file_name, 'r') as file:
            text = file.read()
            unique_chars = set(text)
            return len(unique_chars)
    except FileNotFoundError:
        print("File not found")

file_name = "Textfile.txt" 
print("Number of unique characters in the file:", unique_characters(file_name))
```
